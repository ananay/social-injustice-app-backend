## Calculate Index

### National Index

Just percentage of hateful tweets out of all tweets (keeping it simple)

 0 = Hate Speech
 
 1 = offensive Language
 
 2 = neither

```python
classification_results = df['hate_class'].value_counts()
total_classified = sum(classification_results)
num_hate = classification_results[0]+classification_results[1]

national_index = num_hate / total_classified

national_index, num_hate, total_classified
```

### Per State Index

```python
states = states_map.values()
data = []

for state in states:
    per_state_results = df[df['state']==state]['hate_class'].value_counts()
    
    try:
        state_num_classified = sum(per_state_results)
        state_num_hate = per_state_results[1]
        state_index = state_num_hate / state_num_classified
    except KeyError:
        state_index = 0.0
        state_num_hate = 0
    except ZeroDivisionError:
        state_index = 0.0
        
    data.append([state_index, state_num_hate, state_num_classified])

states_index = pd.DataFrame(data, index=list(states), columns=['index', 'num_hate', 'num_classified'])
states_index
```

```python
states_index['normalized_index'] = states_index['index'].map(lambda x: (x / national_index)*100)

states_index
```
